```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Mall_Customers.csv')
feature = data.iloc[:,[3,4]].to_numpy()
feature
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1,11):
    k_mean = KMeans(n_clusters=i,init='k-means++',random_state=42,n_init=1)
    k_mean.fit(feature)
    wcss.append(k_mean.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Clustering') 
plt.xlabel('Number of clusters')
plt.ylabel('WCSS value')
plt.show()
```

```python
model = KMeans(n_clusters=5,n_init=1,init='k-means++',random_state=42)
label = model.fit_predict(feature)
```

```python
label
```

```python
colors = ['m','y','g','b','k']
for i in range(5):
    plt.scatter(feature[label == i,0],feature[label == i,1],s=40,c=colors[i],label=f'cluster-{i}')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s = 80,c='r',label='Cluster Centers')
plt.title('Clustering Sample')
plt.ylabel('Spending Score')
plt.xlabel('Annual Income')
plt.legend()
plt.show()
```
